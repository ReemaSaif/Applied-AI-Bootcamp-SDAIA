### 1. Antibiotic use in children

The bar plot and the pie chart below show the distribution of pre-existing medical conditions of children involved in a study on the optimal duration of antibiotic use in treatment of tracheitis, which is an upper respiratory infection.

![](https://github.com/HassanAlgoz/B5/blob/main/content/W2/assets/viz_exercise_1.png?raw=1)

a. What features are apparent in the bar plot but not in the pie chart?

b. What features are apparent in the pie chart but not in the bar plot?

c. Which graph would you prefer to use for displaying these categorical data?

# **Answers:**

a. A bar plot makes it easier to see:
* **Exact magnitudes:** You can see and compare the actual lengths of bars, making it easy to judge which categories are the same in value for example Trauma and Neuromuscular.
* **Rank ordering:** is immediately clear which conditions are most and least common for example most common is prematurity and least common is Gastrointestinal and Immunocompromised.
* **Scale reference:** The axis provides a numeric scale, allowing approximate quantitative reading.


b. A pie chart help us in:
* **Overall composition at a glance:** The circular layout highlights proportions of the whole group make up 100%.
* **Relative share of each condition:** You can quickly see which categories dominate the share more for example Prematurity is visually the largest slice.
* **How evenly the data are distributed:** You can see whether the data are fairly spread or heavily concentrated in one or two categories.

c. A bar plot is generally preferred because it allows accurate comparisons between categories, handles many categories better than a pie chart, and remains readable even when some categories are small.


### 2. Black Lives Matter.

A Washington Post-Schar School poll conducted in the United States in June 2020, among a random national sample of 1,006 adults, asked respondents whether they support or oppose protests following George Floyd’s killing that have taken place in cities across the US. The survey also collected information on the age of the respondents. (Washington Post 2020) The results are summarized in the stacked bar plot below.

![](https://github.com/HassanAlgoz/B5/blob/main/content/W2/assets/viz_exercise_3.png?raw=1)

a. Based on the stacked bar plot, do views on the protests and age appear to be associated? Explain your reasoning.

b. Conjecture other possible variables that might explain the potential association between these two variables.

# **Answers:**

a. Yes, views on the protests and age appear to be associated:
* Younger adults (18–29) show higher levels of support, with large portions of the bars in light and dark blue (“Somewhat support” and “Strong”).
* Older adults (65+) show higher levels of opposition, with larger orange and pink segments (“Strongly oppose” and “Somewhat oppose”).

Overall, as age increases, support tends to decrease and opposition tends to increase, meaning opinions change systematically with age rather than remaining roughly constant across all age groups.

b. Several other factors could influence opinions about the protests for example:
* Education level
* Race
* Political affiliation
* Geographic location
* Media consumption habits
* Personal experiences with policing or protests

### 3. Heart transplant data display.

The Stanford University Heart Transplant Study was conducted to determine whether an experimental heart transplant program increased lifespan. Each patient entering the program was officially designated a heart transplant candidate, meaning that they were gravely ill and might benefit from a new heart. Patients were randomly assigned into treatment and control groups. Patients in the treatment group received a transplant, and those in the control group did not. The visualizations below display two different versions of the study results.9 (Turnbull, Brown, and Hu 1974)

![](https://github.com/HassanAlgoz/B5/blob/main/content/W2/assets/viz_exercise_5.png?raw=1)

a. Provide one aspect of the two group comparison that is easier to see from the stacked bar plot (left)?

b. Provide one aspect of the two group comparison that is easeir to see from the standardized bar plot (right)?

c. For the Heart Transplant Study which of those aspects would be more important to display? That is, which bar plot would be better as a data visualization?

# **Answers:**

a. The overall number of patients in each group and how many patients survived or died in each group.

b. The proportions of outcomes within each group and the survival rates regardless of how many patients are in each group.

c. The standardized bar plot that shows proportion is better because the main goal of the study is to compare survival rates between the treatment and control groups, not just the number of patients in each group.

### 4. Meat consumption and life expectancy.

In data collected for You et al. (2022), total meat intake is associated with life expectancy (at birth) in 175 countries. Meat intake is measured in kg per capita per year (averaged over 2011 to 2013). The two ridge plots show an association between income and meat consumption (higher income countries tend to eat more meat) and an association between income and life expectancy (higher income countries have higher life expectancy).

![](https://github.com/HassanAlgoz/B5/blob/main/content/W2/assets/viz_exercise_7.png?raw=1)

a. Do the graphs above demonstrate that meat consumption and life expectancy are associated? That is, can you tell if countries with low meat consumption have low life expectancy? Explain.

b. Let’s assume that you had a plot comparing meat consumption and life expectancy, and they do seem associated. Your friend says that the plot shows that high meat consumption leads to a longer life. You correctly say, no, we can’t tell if there is a causal realtionship because the relationship is confounded by income level. Explain what you mean.

c. How can you investigate the relationship between meat consumption and life expectancy in the presence of confounding variables (like income)?

# **Answers:**

a. No, the graphs do not show a clear association between meat consumption and life expectancy. Instead, they show two separate relationships:
* The top graph shows income is associated with meat consumption
* The bottom graph shows income is also associated with life expectancy

Because both variables are related to income, we cannot conclude from these plots alone that countries with low meat consumption have low life expectancy.



b. A confounder creates a false link between two variables. In this case, income affects both meat consumption and life expectancy. If we don’t account for income, it can seem like meat consumption increases life expectancy, even though income is the real cause.


c. This can be done by comparing countries within the same income group or by using statistical methods such as multiple regression to adjust for income, in order to determine whether meat consumption still predicts life expectancy after accounting for those other factors.

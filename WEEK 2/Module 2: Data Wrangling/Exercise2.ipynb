### Step 13. Turn the item price into a float

#### Step 13.a. Check the item price type

Step 13.b. Create a lambda function and change the type of item price

```python
chipo.item_price = chipo.item_price.apply(lambda x: float(x[1:-1]))
```

### Step 14. How much was the revenue for the period in the dataset?

```python
revenue = (chipo['quantity']* chipo['item_price']).sum()
print("Revenue was = $",round(revenue))
```

### Step 15. How many orders were made in the period?

```python
order = chipo.order_id.value_counts().count()
print("Total orders = ",order)
```

### Step 16. What is the average revenue amount per order?

```python
# Solution 1
revenue_per_order = revenue / chipo.order_id.value_counts().count()
print("Average Revenue Amount Per Order ≈ $",round(revenue_per_order))
```

```python
# Solution 2
chipo['order_revenue'] = chipo.quantity * chipo.item_price
average_revenue = chipo.groupby('order_id').order_revenue.sum().mean()
print("Average Revenue Amount Per Order ≈ $",round(average_revenue))
```

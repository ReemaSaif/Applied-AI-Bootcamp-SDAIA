### Step 1. Import the necessary libraries

```python
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/datasets/investor-flow-of-funds-us/master/data/weekly.csv).

### Step 3. Assign it to a variable called

```python
investor = pd.read_csv("https://raw.githubusercontent.com/datasets/investor-flow-of-funds-us/master/data/weekly.csv")
investor.head()
```

### Step 4.  What is the frequency of the dataset?

* Weekly as we can see that each row corresponds to a one-week interval. Also the file name is ....weekly.csv


### Step 5. Set the column Date as the index.

```python
investor = investor.set_index("Date")
investor.head()
```

### Step 6. What is the type of the index?

```python
investor.index.dtype
```

### Step 7. Set the index to a DatetimeIndex type

```python
investor.index = pd.to_datetime(investor.index)
investor.index.dtype
```

### Step 8.  Change the frequency to monthly, sum the values and assign it to monthly.

```python
monthly = investor.resample('M').sum()
monthly.head()
```

### Step 9. You will notice that it filled the dataFrame with months that don't have any data with NaN. Let's drop these rows.

```python
monthly = monthly.dropna(how='all')
monthly.head()
```

### Step 10. Good, now we have the monthly data. Now change the frequency to year.

```python
yearly = monthly.resample('Y').sum()
yearly.head()
```

### BONUS: Create your own question and answer it.

**Which year had the highest total net inflow across all fund categories?**

```python
high_year = yearly['Total'].idxmax()
print("The highest yearly total was in:", high_year)
```

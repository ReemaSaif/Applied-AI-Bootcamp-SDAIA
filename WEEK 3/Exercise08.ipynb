### Step 2 — Quantify memory *before* and *after* one-hot encoding

In the next few cells you will:

### **Answer:**

   * 46 columns
   * I expect the number of columns to increase by 300.
   * I expect the memory usage to increase by 20x.

```python
ncols_before = cat_data.shape[1]
mem_before = cat_data.memory_usage(deep=True).sum() / 1024 ** 2   # converts bytes to megabytes (MB)

print(ncols_before)    # the number of categorical columns BEFORE encoding
print(mem_before)      # how much memory those columns use BEFORE encoding
```

```python
encoder = OneHotEncoder(
    sparse_output=False,
    handle_unknown='ignore',
)

encoder.set_output(transform='pandas')
```

```python
# transforms the categorical data into one‑hot encoded columns

cat_data_encoded = encoder.fit_transform(cat_data)
```

```python
ncols_after = cat_data_encoded.shape[1]
mem_after = cat_data_encoded.memory_usage(deep=True).sum() / 1024**2  # MB
```

```python
print(ncols_after)    # the number of categorical columns AFTER encoding
print(mem_after)      # how much memory those columns use AFTER encoding
```
